```python
# import dependencies
import pandas as pd
from pathlib import Path
```

```python
# reference file path
data = Path("../data/GamingStudy_data.csv")

# import data into dataframe
data_df = pd.read_csv(data, encoding="windows-1252")

# display DataFrame
data_df.head()
```

```python
# remove unused columns
clean_df = data_df.drop(columns = [
    'Timestamp', 'Platform', 'League', 'highestleague', 'streams', 'Narcissism', 
    'Birthplace', 'Residence', 'Reference', 'accept', 'earnings', 'GADE'
    ])

# rename columns
clean_df = clean_df.rename(columns = {
    'S. No.': 'Entry', 'Residence_ISO3': 'Residence', 'Birthplace_ISO3': 'Birthplace', 'whyplay': 'Gaming Motivation', 
    'Hours': 'Hours per Week', 'GAD_T': 'GAD Total', 'SWL_T' : 'SWL Total', 'SPIN_T': 'SPIN Total'
    })
# reorder columns
clean_df = clean_df[[
    'Entry', 'Gender', 'Age', 'Residence', 'Birthplace', 'Work', 'Degree', 'Game', 'Hours per Week', 'Gaming Motivation', 'Playstyle',
    'GAD1', 'GAD2', 'GAD3', 'GAD4', 'GAD5', 'GAD6', 'GAD7', 'GAD Total',
    'SWL1', 'SWL2', 'SWL3', 'SWL4', 'SWL5', 'SWL Total',
    'SPIN1', 'SPIN2', 'SPIN3', 'SPIN4', 'SPIN5', 'SPIN6', 'SPIN7', 'SPIN8', 'SPIN9', 
    'SPIN10', 'SPIN11', 'SPIN12', 'SPIN13', 'SPIN14', 'SPIN15', 'SPIN16', 'SPIN17', 'SPIN Total'
    ]]

# drop rows with NA
clean_df = clean_df.dropna()

# display DataFrame
clean_df.head(10)
```

```python
# change floats to integers
clean_df[[
    'Hours per Week', 
    'SPIN1', 'SPIN2', 'SPIN3', 'SPIN4', 'SPIN5', 'SPIN6', 'SPIN7', 'SPIN8', 'SPIN9', 'SPIN10', 
    'SPIN11', 'SPIN12', 'SPIN13', 'SPIN14', 'SPIN15', 'SPIN16', 'SPIN17', 'SPIN Total']] = \
clean_df[[
    'Hours per Week', 'SPIN1', 'SPIN2', 'SPIN3', 'SPIN4', 'SPIN5', 'SPIN6', 'SPIN7', 'SPIN8', 
    'SPIN9', 'SPIN10', 'SPIN11', 'SPIN12', 'SPIN13', 'SPIN14', 'SPIN15', 'SPIN16', 'SPIN17', 'SPIN Total'
        ]].astype(int)

# display DataFrame
clean_df.head()
```
